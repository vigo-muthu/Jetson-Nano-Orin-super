```python
import cv2 as cv
from ultralytics import YOLO
import ipywidgets as widgets
from IPython.display import display

# FORCE SMALL MODEL FOR SPEED
model = YOLO("yolov8n.pt")

img_widget = widgets.Image(format='jpeg')
display(img_widget)

cap = cv.VideoCapture(0, cv.CAP_V4L2)
cap.set(cv.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv.CAP_PROP_FRAME_HEIGHT, 480)

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # =========================
    # SPEED FIX: RUN YOLO EVERY 3 FRAMES
    # =========================
    frame_count += 1

    if frame_count % 3 == 0:
        results = model.predict(frame, conf=0.4, device=0, verbose=False)
        annotated = results[0].plot()

        _, buffer = cv.imencode('.jpg', annotated)
        img_widget.value = buffer.tobytes()

cap.release()
```
